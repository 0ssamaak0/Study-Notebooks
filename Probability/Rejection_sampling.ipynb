```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

```python
def makeSample():
    flu = stats.bernoulli.rvs(0.1)
    und = stats.bernoulli.rvs(0.8)

    if flu == 1: 
        fev = stats.norm.rvs(100, 1.81)
    else:
        fev = stats.norm.rvs(98.25, 0.73)
    fev = round(fev, 2)
    if und == 1 and flu == 1:
        tir = stats.bernoulli.rvs(1)
    elif und == 1 and flu == 0:
        tir = stats.bernoulli.rvs(0.8)
    elif und == 0 and flu == 1:
        tir = stats.bernoulli.rvs(0.9)
    elif und == 0 and flu == 0:
        tir = stats.bernoulli.rvs(0.1)

    return [flu, und, fev, tir]
```

```python
def matches_pattern(sample, pattern):
    """
    Checks if a sample matches a pattern.
    pattern: list with values or None (None means 'any')
    """
    return all(p is None or s == p for s, p in zip(sample, pattern))

def rejection_sampling(sample_func, pattern, num_samples):
    accepted = 0
    total = 0
    from tqdm import tqdm
    pbar = tqdm(range(num_samples), desc="Total samples")
    for _ in pbar:
        sample = sample_func()
        total += 1
        if matches_pattern(sample, pattern):
            accepted += 1
            if accepted == num_samples:
                break
    pbar.close()
    return accepted / total

# Example usage: estimate P([x, 1, x, 0]) where x can be any value
pattern = [None, 1, round(100, 2), 1]
prob = rejection_sampling(makeSample, pattern, 10000)
print(prob)
```
