```python
import requests
import json
import chromadb

chroma_client = chromadb.PersistentClient(path="db/")
```

```python
def get_embeddings(text):
    url = "http://localhost:11434/api/embeddings"
    data = {"model": "all-minilm:latest", "prompt": f"{text}"}
    response = requests.post(url, data=json.dumps(data))

    # If you want to print the response
    return response.json()["embedding"]
```

```python
collection = chroma_client.create_collection(name="my_collection")
```

```python
documents = [
    "llama are members of the camelid family meaning they're pretty closely related to vicu√±as and camels",
    "llama were first domesticated and used as pack animals 4,000 to 5,000 years ago in the Peruvian highlands",
    "llama can grow as much as 6 feet tall though the average llama between 5 feet 6 inches and 5 feet 9 inches tall",
    "llama weigh between 280 and 450 pounds and can carry 25 to 30 percent of their body weight",
    "llama are vegetarians and have very efficient digestive systems",
    "llama live to be about 20 years old, though some only live for 15 years and others live to be 30 years old",
]
```

```python
for i, doc in enumerate(documents):
    embeddings = get_embeddings(doc)
    new_messy.upsert(
        ids=str(i), embeddings=embeddings, documents=doc, metadatas={"source": "messy"}
    )
```
